```python
country_list = ["Belgium","England","France","Germany","Italy","Netherlands","Portugal","Spain","Greece","Scotland"]
def get_win_perc(country_name):
    temp_df = pd.read_csv("data/Normalized_Features/"+country_name+"_features.csv")
    print(country_name, len(temp_df.loc[temp_df['win'] == 1])/len(temp_df))

for country in country_list:
    get_win_perc(country)
```
