```python
import requests
from bs4 import BeautifulSoup
```

```python
def get_(search_string):
    final_string = ""
    for i in search_string.split():
        final_string = final_string + i + '_'
    return final_string[:-1]



def get_table_data(search_query):
    url = "https://en.wikipedia.org/wiki/" + search_query
    page = requests.get(url)
    soup = BeautifulSoup(page.content, 'html.parser')
    table = soup.find_all("table")
    nation = ""
    for i in soup.find_all("table", {"class": "infobox vcard"}):
        for j in i.find_all("td", {"class": "infobox-data birthplace"}):
            if i is not None and j is not None:
                nation = j.text.split(", ")[-1]

    return nation
```

```python
final_df = pd.read_csv("created/shortened.csv")
final_df["Nationality"] = np.nan

for data in final_df["Name"].iteritems():
    query_string = get_(data[1])
    final_df.iloc[data[0], 12] = get_table_data(query_string)
    print(data[0])
```
